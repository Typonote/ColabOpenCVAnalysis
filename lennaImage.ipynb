# **이미지 위에 정보 표시하기**

**이미지 위에 선 그리기**

cv2.line(image, start_point, end_point, color, line_thickness, line_type) <br>

*   start_point, end_point: 튜플로 구성된 x, y 좌표로 행렬 좌표가 아님
*   color : 선의 색상을 지정. (B, G, R)로 구성되어 있으며0~ 255 사이값을 사용
*   line_thickness: 선의 굵기
*   line_type: 선의 종류

cv2.circle(image, center_point, radius, color, line_thickness, line_type)


*   center_point: 원의 중심점
*   radius : 반지름 길이



```python
# 줄 그리기
line_img = cv2.line(image, (0,0),(image.shape[1],image.shape[0]), (255,0,0),1,cv2.LINE_AA)
# 원 그리기
round_img = cv2.circle(image, (image.shape[0] // 2 ,image.shape[1] // 2), 100,(0,255,0),5)
```

```python
# 저장
cv2.imwrite('/content/drive/MyDrive/lenaImage/images/LineImage.jpg', line_img)
cv2.imwrite('/content/drive/MyDrive/lenaImage/images/CircleImage.jpg', round_img)
```

```python
# 저장된 이미지 읽어오기
aa = cv2.imread('/content/drive/MyDrive/lenaImage/images/LineImage.jpg')
cv2_imshow(aa)
```

**이미지 위에 텍스트 작성하기**

cv2.putText(image, string, text_bottom_left, font_type, font_scale, color, line_thickeness, line_type) <br>



*   string: 이미지에 쓸 문자열로 영문만 쓸 수 있음
*   text_bottom_left: 문자열이 시작되는 좌측 하단 좌표



```python
# 텍스트 작성하기
text_img = cv2.putText(image,'This is lenna',(100,100),cv2.FONT_HERSHEY_SIMPLEX,0.7,(128,0,128))
```

```python
cv2.imwrite('/content/drive/MyDrive/lenaImage/images/TextImage.jpg', text_img)
```
